```python
#menghapus tanda baca
import string

print(string.punctuation)
sentence_baru = "halo! semua apa kabar?"
```

```python
#using regex
regex_punc = re.compile('[%s]' % re.escape(string.punctuation)) #membuat sebuah patern dalam regex
new_sentence3 = regex_punc.sub('', sentence_baru)
new_sentence3
```

```python
#using translate
new_sentence2 = sentence_baru.translate(str.maketrans('', '', string.punctuation))
print("kalimat tanpa tanda baca dengan translate:", new_sentence2)
```

```python
#regex patern
new_sentence3 = re.sub(r"[^\w\d\s]+", "",sentence_baru) #apabila textku, ada sesuatu yang bukan (^) \w = alphanumeric, \d = digit, \s = spasi
new_sentence3
```

```python
def change_director_name(name):
  name = name.lower()
  new_name = name.replace("scott", "skubidu")
  return new_name

df_imdb['new_director'] = df_imdb['Director'].apply(change_director_name)
df_imdb.head(5)
```

```python
#remove emoji
sentence_emoji = "ðŸ™‚ halo semua ðŸ™‚"

emoji_pattern = re.compile("["
                               u"\U0001F600-\U0001F64F"  # emoticons
                               u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                               u"\U0001F680-\U0001F6FF"  # transport & map symbols
                               u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                               u"\U00002702-\U000027B0"
                               u"\U000024C2-\U0001F251"
                               "]+", flags=re.UNICODE)
print(emoji_pattern.sub(r'', sentence_emoji))
```

```python
#menghitung jumlah film berdasarkan genre

genre_count = df_imdb["Genre"].apply(lambda x: x.split(',')).apply(lambda x: pd.Series(x)).stack().value_counts()
genre_count = pd.DataFrame({"Genre":genre_count.index,"Total":genre_count.values})
genre_count

#apply(lambda x: x.split(',')) -> change genre with str type to list -> become series
#apply(lambda x: pd.Series(x)) -> change each row into series -> become dataframe
#stack() -> make all series into one series
#value_count() -> count total number of each genre
```

```python
sns.barplot(data=genre_count[:5],x="Genre",y="Total")
```

```python
#pie chart

palette_color = sns.color_palette('pastel')
plt.pie(genre_count['Total'][:5], labels=genre_count['Genre'][:5], colors=palette_color, autopct='%1.0f%%', pctdistance=0.5, labeldistance=1.2)
```

```python
#line chart / diagram garis

year_series = df_imdb.groupby('Year').size() #menghitung jumlah tiap unique value dalam suatu kolom
df_year = pd.DataFrame({'year':year_series.index,'total_movie':year_series.values})
df_year
```

```python
sns.lineplot(data=df_year, x="year", y="total_movie",marker="o")
plt.grid()
```

```python
# BESARAN RATA-RATA GAJI DATA SCIENTIST DALAM 5 TAHUN TERAKHIR DARI 2 PROVINSI YANG BERBEDA

dict_salary = {
    "year":[2021,2022,2023,2024,2025,2021,2022,2023,2024,2025],
    "province":["DKI","DKI","DKI","DKI","DKI","JABAR","JABAR","JABAR","JABAR","JABAR"],
    "salary":[6,7,8.5,9,10,7,7.2,8.1,9.5,11]
}

df_salary = pd.DataFrame({'year':dict_salary['year'],'province':dict_salary['province'],'salary':dict_salary['salary']})
df_salary
```

```python
x_year = df_salary['year'][:5].to_list() #membuat list dari dataframe kolom year
y_dki = df_salary['salary'][df_salary['province'] == "DKI"].to_list() #membuat list dari salary dengan provinsi = DKI
y_jabar = df_salary['salary'][df_salary['province'] == "JABAR"].to_list() #membuat list dari salary dengan provinsi = JABAR

#using matplotlib
plt.plot(x_year, y_dki, label = "DKI", marker="o")
plt.plot(x_year,y_jabar, label = "JABAR", marker="o")
plt.legend()
plt.grid()
plt.show()
```

```python
#using seaborn
sns.lineplot(x=x_year, y=y_dki,marker="o")
sns.lineplot(x=x_year, y=y_jabar,marker="o")
```
